```python
import pandas as pd
import os
import re
```

```python
date_formats = {
    'season-0910_csv.csv': '%Y-%m-%d',
    'season-1011_csv.csv': '%d/%m/%y',
    'season-1112_csv.csv': '%d/%m/%y',
    'season-1213_csv.csv': '%d/%m/%y',
    'season-1314_csv.csv': '%d/%m/%y',
    'season-1415_csv.csv': '%d/%m/%y',
    'season-1516_csv.csv': '%d/%m/%y',
    'season-1617_csv.csv': '%d/%m/%y',
    'season-1718_csv.csv': '%d/%m/%y',
    'season-1819_csv.csv': '%d/%m/%Y',
    'season-1920_csv.csv': '%d/%m/%Y',
}
```

```python
# Consolidating all the seasons data
epl_data = pd.DataFrame()

for filename in os.listdir('../Data'):
    if filename.endswith('csv') and filename.startswith('season'):
        print(filename)
        season_data = pd.read_csv('../Data/'+filename)
        season_data['Season'] = 'S'+re.findall('\d\d\d\d', filename)[0]
        season_data.Date = pd.to_datetime(season_data.Date, format=date_formats[filename])
        epl_data = pd.concat([epl_data, season_data])
        
epl_data.to_csv('../Data/EPL_Data_Full.csv', index=False)
        
print(epl_data.shape)
epl_data.head()
```
